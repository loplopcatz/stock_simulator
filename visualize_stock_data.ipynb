```python
# === PHASE 1: STOCK DATA VISUALIZATION ===
# 📊 Goal: Explore TSLA stock trends visually using Pandas & Matplotlib
# 🧠 Skills Practiced: EDA, Data Cleaning, Visualization
# 🎯 Learning Goals: Understand stock movements, prep for modeling, build readable charts

# === 1. IMPORT LIBRARIES ===
import pandas as pd
import matplotlib.pyplot as plt

# === 2. LOAD CLEANED DATA ===
# 📦 Input file created earlier using Alpha Vantage API
# 🔐 Make sure it's the cleaned version with float-converted columns
file_path = "TSLA_stock_data.csv"
df = pd.read_csv(file_path, parse_dates=['date'], index_col='date')

# === 3. PLOT LINE CHART: CLOSING PRICE OVER TIME ===
# 📈 This chart shows the overall trend of Tesla's closing prices over time.
# Useful to understand the general movement, spikes, or drops in value.
plt.figure(figsize=(14, 6))
plt.plot(df['close'], label='Closing Price', linewidth=2)
plt.title('Tesla (TSLA) Stock Closing Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 📝 Chart Summary:
# This chart provides a visual overview of Tesla's stock price closing trends.
# Observe any consistent upward or downward trends, and identify periods of volatility.
```

```python
# === 4. PLOT MOVING AVERAGES (MA10, MA30) ===
# 📝 Chart Summary:
# The moving average overlays reveal smoothed price trends.
# Crosses between MA10 and MA30 often signal potential momentum changes.
# 📉 This chart compares the closing price to its 10-day and 30-day moving averages.
# Helps visualize momentum shifts and support/resistance zones.
df['MA10'] = df['close'].rolling(window=10).mean()
df['MA30'] = df['close'].rolling(window=30).mean()

plt.figure(figsize=(14, 6))
plt.plot(df['close'], label='Close', alpha=0.5)
plt.plot(df['MA10'], label='10-day MA', linestyle='--')
plt.plot(df['MA30'], label='30-day MA', linestyle='--')
plt.title('TSLA Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# 📝 Chart Summary:
# This chart provides a visual overview of Tesla's stock price closing trends.
# Observe any consistent upward or downward trends, and identify periods of volatility.
```

```python
# === 5. HIGHLIGHT MAXIMUM AND MINIMUM CLOSE ===
# 📝 Chart Summary:
# This chart helps highlight the range and extremes in Tesla's stock prices.
# Green marker = historical high, Red marker = historical low — helpful for spotting turning points.
# 📍 This chart highlights the highest and lowest closing prices with colored markers.
# Useful for understanding the stock's volatility and historical range.
max_date = df['close'].idxmax()
min_date = df['close'].idxmin()

plt.figure(figsize=(14, 6))
plt.plot(df['close'], label='Close', linewidth=2)
plt.scatter(max_date, df['close'].max(), color='green', label='Max', s=100, zorder=5)
plt.scatter(min_date, df['close'].min(), color='red', label='Min', s=100, zorder=5)
plt.title('TSLA: Highs and Lows in Closing Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# 📝 Chart Summary:
# This chart provides a visual overview of Tesla's stock price closing trends.
# Observe any consistent upward or downward trends, and identify periods of volatility.
```
